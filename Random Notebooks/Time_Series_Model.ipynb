```python
import torch
from torch.utils.data import Dataset, DataLoader
import os
import torch.nn as nn
```

```python
name_list = None

for root, dirs, files in os.walk(".", topdown = False):
  name_list = files
```

```python
name_list = [i for i in name_list if i[-1] == 't']
```

```python
name_list
```

```python
char_set = set()

num_classes = len(name_list)
data_set = []
class_index = 0
for i in name_list:
  with open(i, "r") as file:
    temp_list = [i.strip() for i in file.readlines()]



    for word in temp_list:
      for char in word:
        char_set.add(char)
      data_set.append((word, class_index))
    class_index +=1


```

```python
print(data_set)
```

```python
char_index_dict = dict()

char_list = list(char_set)

for index, val in enumerate(char_set):
  char_index_dict[val] = index
```

```python
print(char_index_dict)
```

```python
def word_to_tensor(word):
  word_tensor = torch.zeros(len(word), 1, len(char_list))

  for index, val in enumerate(word):
    char_index = char_index_dict[val]

    word_tensor[index, 0, char_index] = 1

  return word_tensor
```

```python
word_to_tensor('word')
```

```python
class CharDataset(Dataset):
  
  def __init__(self, data_Set):
    self.data_set = data_set
  
  def __len__(self):
    return len(self.data_set)

  def __getitem__(self, index):
    return self.data_set[index]
```

```python
DS = CharDataset(data_set)
```

```python
train_size = int(0.8 * len(DS))
test_size = len(DS) - train_size

train_dataset, test_dataset = torch.utils.data.random_split(DS, [train_size, test_size])
```

```python
class RNN(nn.Module):
  
  def __init__(self, char_size, hidden_size, output_size):
    super(RNN, self).__init__()
    self.hidden = nn.Linear(char_size + hidden_size, hidden_size)
    self.output = nn.Linear(char_size + hidden_size, output_size)
    self.tanh = nn.Tanh()
    self.hidden_size = hidden_size

  def forward(self, x, hidden):
    concat = torch.cat((x, hidden), 1)
    hiddne = self.hidden(concat)
    hidden = self.tanh(hidden)
    output = self.output(concat)
    return output, hidden

  def initHidden(self):
    return torch.zeros(1, self.hidden_size)
```

```python
Pred_Model = RNN(len(char_list), num_classes, num_classes)
```

```python
criterion = nn.CrossEntropyLoss()
optim  = torch.optim.Adam(Pred_Model.parameters(), lr=1e-3)
```

```python
epochs = 10

for i in range(epochs):
  print("Epoch", i + 1)

  for index, val in enumerate(train_dataset):
    word, cls = val
    cls = torch.tensor([cls])
    word_tensor = word_to_tensor(word)
    hidden = Pred_Model.initHidden()
    output = None

    for i in range(len(word)):
      output, hidden = Pred_Model(word_tensor[i], hidden)

    optim.zero_grad()
    loss = criterion(output, cls)
    loss.backward()
    optim.step()

```

```python
correct= 0
count = 0

with torch.no_grad():
  for _, val in enumerate(test_dataset):
    count += 1
    word, cls = val
    word_tensor = word_to_tensor(word)
    hidden = Pred_Model.initHidden()
    output = None

    for i in range(len(word)):
      output, hidden = Pred_Model(word_tensor[i], hidden)

    predict_class = torch.argmax(output).item()
    if predict_class == cls:
      correct += 1

print(correct/count)
```
